<h1>PART-1</h1>

```python
import pandas as pd

tickets = pd.read_csv("data_safe_copy.csv")

tickets_c = tickets.copy()

print(tickets_c.shape[0])

tickets_c.head()

tickets_c.info()
```

<h1>PART-2</h1>

```python
tickets_c["opened_at"] = pd.to_datetime(tickets_c["opened_at"])

tickets_c.info()
```

```python
tickets_c["category"] = tickets_c["category"].str.strip()
tickets_c["category"] = tickets_c["category"].str.lower()
tickets_c["priority"] = tickets_c["priority"].str.strip()
tickets_c["priority"] = tickets_c["priority"].str.lower()

print(tickets_c["category"].unique())
print(tickets_c["priority"].unique())


```

<h1>PART-3</h1>

```python
invalids = tickets_c["resolution_minutes"].isna()

unknows = tickets_c["resolution_minutes"] == "unknown"

invalids = invalids | unknows

tickets_c["resolution_minutes"] = pd.to_numeric(tickets_c["resolution_minutes"],errors="coerce")

missing_values = tickets_c["resolution_minutes"].isna()

bool (missing_values.sum() == invalids.sum())
```

<h1>PART-4</h1>

```python
tickets_clean = tickets_c.dropna(subset=["resolution_minutes"])

tickets_clean = pd.DataFrame(tickets_clean)

tickets_clean


non_negative = all(tickets_clean["resolution_minutes"] >= 0)

all(tickets_clean["category"] == tickets_clean["category"].str.lower())
all(tickets_clean["priority"] == tickets_clean["priority"].str.lower())

print(non_negative == True)

print(all(tickets_clean["category"] == tickets_clean["category"].str.lower()))
print(all(tickets_clean["priority"] == tickets_clean["priority"].str.lower()))

tickets_clean.shape[0]
tickets_c.shape[0]
```

<h1>PART-5</h1>

```python
summary = {
    "total_records": tickets_c.shape[0] ,
    "clean_records": tickets_clean.shape[0],
    "missing_resolution":int(missing_values.sum()),
    "avg_resolution":float(tickets_clean["resolution_minutes"].mean()),
    "max_resolution":int(tickets_clean["resolution_minutes"].max())
    }

print(summary)

summary["clean_records"] + summary["missing_resolution"] == summary["total_records"]
```
